```python
from pycocotools.coco import COCO
import os
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt

dataroot = r"D:\\\coco2017"

# instanceCOCO API
dataDir = dataroot
dataType = 'train2017'
instances_annFile = os.path.join(dataDir, 'annotations_trainval2017/instances_{}.json'.format(dataType))
coco = COCO(instances_annFile)

# image captionCOCO API
captions_annFile = os.path.join(dataDir, 'annotations_trainval2017/captions_{}.json'.format(dataType))
coco_caps = COCO(captions_annFile)

# ID
ids = list(coco.anns.keys())
catIds = coco.getCatIds(catNms=['person', 'dog', 'skateboard'])
imgIds = coco.getImgIds(catIds=catIds)
imgIds = coco.getImgIds(imgIds=[324158])


# URL
ann_id = np.random.choice(ids)
img_id = coco.anns[ann_id]['image_id']
img = coco.loadImgs(img_id)[0]
# loadImgs(), [0]
# , : ["license", "file_name", "coco_url", "height", "width", "date_captured", "id"]
url = img['coco_url']

# print URL and visualize corresponding image
print(url)
I = io.imread(url)
plt.axis('off')
plt.imshow(I)
plt.show()

# load and display captions
annIds = coco_caps.getAnnIds(imgIds=img['id']);
anns = coco_caps.loadAnns(annIds)
coco_caps.showAnns(anns)
```
